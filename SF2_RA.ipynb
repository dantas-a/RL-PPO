```python
# Permet de jouer au jeu
import retro
# Permet de ralentir la vitesse 
import time
```

## Jouer au jeu en tirant des actions aléatoires

```python
#Lance l'environnement du jeu
env = retro.make(game='StreetFighterIISpecialChampionEdition-Genesis')
```

```python
# Remet le jeu à zéro
obs = env.reset()
# Partie non terminée
done = False
# Une partie
for game in range(1) :
    #Tant qu'on a pas perdu
    while not done:
        #Si on termine la partie, remet le jeu à zéro
        if done:
            obs = env.reset()
        #Affiche le jeu
        env.render()
        #Joue une action aléatoire et récupère les paramètres
        obs, reward, done, info = env.step(env.action_space.sample())
        #Pause entre chaque action
        time.sleep(0.01)
        print(info)
        #Affiche la récompense
        print(reward)
```

```python
#Ferme l'environnement du jeu
env.close()
```
